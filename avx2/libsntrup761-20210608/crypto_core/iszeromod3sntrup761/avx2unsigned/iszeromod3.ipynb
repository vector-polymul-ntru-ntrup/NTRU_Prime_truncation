```python
def mulsub3( in_la , m , rase_deg , lb ):
    la = list(in_la)
    if( len(la) < rase_deg + len(lb) ): la = la + [0]*(len(lb)+rase_deg-len(la))
    for i in range(len(lb)):
        la[rase_deg + i] = ((6+la[rase_deg + i])-(m*lb[i]))%3
    return la
        
def muladd3( in_la , m , rase_deg , lb ):
    la = list(in_la)
    if( len(la) < rase_deg + len(lb) ): la = la + [0]*(len(lb)+rase_deg-len(la))
    for i in range(len(lb)):
        la[rase_deg + i] = (la[rase_deg + i] + m*lb[i])%3
    return la

def add3( la, lb ):
    if(len(lb)>len(la)): la,lb = lb,la
    r = list(la)
    for i in range(len(lb)):
        r[i] = (r[i]+lb[i]) %3
    return r

def polymul( la , lb ):
    r = [0]
    for i in range(len(la)):
        r = muladd3(r , la[i] , i , lb)
    return r
    

def div3( dividend , divisor ):
    #print( dividend)
    quo = []
    rem = list(dividend)
    for i in range( len(dividend)-len(divisor) , -1 , -1 ):
        quo.insert(0,rem[len(divisor)+i-1])
        rem = mulsub3( rem , rem[len(divisor)+i-1] , i , divisor )
    return quo, rem[:len(divisor)-1]

def is_vec_zero( vec ):
    r = 0
    for i in vec: r = r | i
    return 0==r
    
        
# tester    
a = [ 1 , 2 ,1 ]
b = [ 1 , 1 ]
quo , rem = div3( a , b )
print(quo)
print(rem)
print(a)
print( polymul(quo,b) )

print(is_vec_zero(rem))
```
